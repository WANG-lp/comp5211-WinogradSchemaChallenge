```python
from xml.dom import minidom
```

```python
xmldoc = minidom.parse('../datasets/WSCollection.xml')
```

```python
schemas = xmldoc.getElementsByTagName('schema')
```

```python
def getStruct(s):
    text = []
    all_t = s.getElementsByTagName('text')
    text.append(all_t[0].getElementsByTagName('txt1')[0].firstChild.data)
    text.append(all_t[0].getElementsByTagName('pron')[0].firstChild.data)
    text.append(all_t[0].getElementsByTagName('txt2')[0].firstChild.data)
    quote = []
    quote_a = s.getElementsByTagName('quote')
    if(len(quote_a[0].getElementsByTagName('quote1')):
        quote.append(quote_a[0].getElementsByTagName('quote1')[0].firstChild.data)
    else:
       quote.append('')
    quote.append(quote_a[0].getElementsByTagName('pron')[0].firstChild.data)
    quote.append(quote_a[0].getElementsByTagName('quote2')[0].firstChild.data)
    return text, quote
    
```

```python
for s in schemas:
    t, q = getStruct(s)
    print '======'
    print t
    print q
    print '======'
    
```
